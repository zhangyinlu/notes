## 梯度下降

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 单变量梯度下降

```python
def target_function(x):
    return x * x

def derivative_function(x):
    return 2 * x

def draw(X):
    x = np.linspace(-1.5, 1.5)
    y = target_function(x)
    plt.plot(x, y)

    Y = []
    for i in range(len(X)):
        Y.append(target_function(X[i]))
    plt.plot(X,Y)

x = 1.2
eta = 0.3
error = 1e-5
X = []
while True:
    y = target_function(x)
    X.append(x)
    print("x=%f,y=%f"%(x,y))
    if(y < error):
        break
    x = x - eta * derivative_function(x)

draw(X)
```

#### 双变量梯度下降

```python
from mpl_toolkits.mplot3d import Axes3D

def target_function_double(x, y):
    return x**2 + np.sin(y)**2

def derivative_function_double(x, y):
    return 2*x, 2*np.sin(y)*np.cos(y)

x = -1.5
y = 1.2
eta = 0.3
error = 1e-5

X = []
Y = []
Z = []

while True:
    z = target_function_double(x, y)
    if(z < error):
        break        
    X.append(x)
    Y.append(y)
    Z.append(z)
    dx,dy = derivative_function_double(x, y)
    print("x=%f,y=%f,z=%f,dx=%f,dy=%f"%(x,y,z,dx,dy))
    x = x - eta * dx
    y = y - eta * dy
```

```python
fig = plt.figure()
ax = Axes3D(fig)
U,V = np.meshgrid(np.linspace(-1.5, 1.5, 100), np.linspace(-1.5, 1.5, 100))
R = U**2 + np.sin(V)**2
ax.plot_surface(U, V, R, cmap='rainbow')
plt.plot(X, Y, Z, color='red')
```

#### 学习率的影响

```python
def draw(X):
    x = np.linspace(-1.5, 1.5)
    y = target_function(x)
    plt.plot(x, y)

    Y = []
    for i in range(len(X)):
        Y.append(target_function(X[i]))
    plt.plot(X,Y)

def LearningRate(eta):
    x = 1.2
    error = 1e-5
    X = []
    for i in range(50):
        X.append(x)
        y = target_function(x)
        if(y < error):
            break
        x = x - eta * derivative_function(x)
    draw(X)
```
