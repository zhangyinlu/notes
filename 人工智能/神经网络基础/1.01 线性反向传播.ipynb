#### 线性反向传播

```python
import numpy as np
import matplotlib.pyplot as plt

w = 3
b = 4
t = 150
error = 1e-5
```

```python
def target_function(w,b):
    x = 2*w + 3*b
    y = 2*b + 1
    z = x*y
    return x,y,z

print(target_function(w,b))
```

```python
def single_variable_w(w,b,t,error):
    print("single variable: w ------------ ")
    while(True):
        x,y,z = target_function(w,b)
        delta_z = z - t
        print("w=%f,b=%f,z=%f,delta_z=%f"%(w,b,z,delta_z))
        if abs(delta_z) < error:
            break
        factor_w = 2*y
        delta_w = delta_z/factor_w
        print("factor_w=%f,delta_w=%f"%(factor_w,delta_w))
        w = w - delta_w
    print("final w=%f" % w)

single_variable_w(w,b,t,error)
```

```python
def single_variable_b(w,b,t,error):
    print("single variable: b ------------ ")
    while(True):
        x,y,z = target_function(w,b)
        delta_z = z - t
        print("w=%f,b=%f,z=%f,delta_z=%f"%(w,b,z,delta_z))
        if abs(delta_z) < error:
            break
        factor_b = 3*y + 2*x
        delta_b = delta_z/factor_b
        print("factor_w=%f,delta_w=%f"%(factor_b,delta_b))
        b = b - delta_b
    print("final b=%f" % b)

single_variable_b(w,b,t,error)
```

```python
def double_variable(w,b,t,error):
    print("double variable: w, b ------------ ")
    while(True):
        x,y,z = target_function(w,b)
        delta_z = z - t
        print("w=%f,b=%f,z=%f,delta_z=%f"%(w,b,z,delta_z))
        if abs(delta_z) < error:
            break
        factor_w = 2*y
        factor_b = 3*y + 2*x
        delta_w = delta_z/factor_w/2
        delta_b = delta_z/factor_b/2
        print("factor_w=%f,factor_b=%f,delta_w=%f,delta_b=%f"%(factor_w,factor_b,delta_w,delta_b))
        b = b - delta_b
        w = w - delta_w
    print("final w=%f" % w)
    print("final b=%f" % b)

double_variable(w,b,t,error)
```
